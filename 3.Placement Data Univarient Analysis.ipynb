#### Explanation:

* **Purpose:**
  The function `quanQual` is designed to separate the columns of a dataset (like a Pandas DataFrame) into two types:

  * **Quantitative (`quan`)**: Numeric data
  * **Qualitative (`qual`)**: Categorical data

* **Step-by-step:**

  1. Takes a DataFrame `dataset` as input.
  2. Initializes two empty lists: `quan` and `qual`.
  3. Iterates over each column name.
  4. Checks the data type of each column:

     * If the type is `'O'` (which means `object` in pandas, commonly used for strings/categories), it's **qualitative**.
     * Otherwise, it's considered **quantitative**.
  5. Appends the column names to the appropriate list.
  6. Returns both lists.

---

````python
def quanQual(dataset):
    quan=[]
    qual=[]
    for columnName in dataset.columns:
        #print(columnName)
        if(dataset[columnName].dtype=='O'):
            #print("qual")
            qual.append(columnName)
        else:
            #print("quan")
            quan.append(columnName)
    return quan,qual


    
````

### 📘 **Understanding Percentiles in Data Analysis**

#### 🔢 What is a Percentile?

A **percentile** is a measure used in statistics to indicate the value below which a given percentage of observations in a group falls or value exists within the range.

It dividing in parts of intial vale,25%(Q1),50%(Q2),75%(Q3),100%(Q1)Final Value.

#### 🔢 What is a Purpose of Percentile?
* Sort the data.
* Asending the order of the data.

#### 🔢Formula: Percentile (i)=k/100*(n+1) (Position of data)

For example:

* The **25th percentile** is the value below which 25% of the data lies.
* The **50th percentile** is the **median**.
* The **90th percentile** is the value below which 90% of the data lies — only the top 10% exceed it.

---

#### 🧠 **Why Are Percentiles Useful?**

Percentiles help in:

* Understanding **data distribution**
* Identifying **outliers**
* Comparing **individual values** against the whole dataset (e.g., "This student scored higher than 80% of their peers")

---

#### 🧮 **How to Calculate Percentiles in Python**

```python
import numpy as np

# Example: Calculate the 25th, 50th, and 75th percentiles of a column

```

---

#### 📏 **Percentile vs Percentile Rank**

* **Percentile Value**: Given a percentile (e.g., 90th), find the score at that point.
* **Percentile Rank**: Given a score, find what percentile it belongs to.

---

#### 📈 **Use Case in Student Data Analysis**

By calculating percentiles for academic percentages and salaries, we can:

* Identify **average vs top performers**

## 1. Interquartile Range (IQR)

## 2. 1.5 × IQR Rule (Outlier Detection)
- **Definition**: Values below `Q1 - 1.5IQR` or above `Q3 + 1.5IQR` are potential outliers.
- **Calculated Ranges**:
  - **Lower Bound (Q1 - 1.5IQR)**:
    - `ssc_p`: 37.95

    - `degree_p`: 44.5

    - `etest_p`: 118.75

## 3. Min and Max Values
- **Min**: The smallest observed value in the dataset.
  - Example: `salary` starts at 200,000 (no values below this).
- **Max**: The largest observed value in the dataset.
  - Example: `salary` peaks at 940,000 (exceeds upper bound → **potential outlier**).

````python
def Univariate(dataset,quan):
    descriptive=pd.DataFrame(index=["Mean","Median","Mode","Min","Max","25%","50%","75%","100%","IQR",
                               "1.5IQR","LesserRange","GreaterRange","Skewness","Kurtosis"],columns=quan)
    
    for columnName in quan:
        descriptive[columnName]["Mean"]=dataset[columnName].mean()
        descriptive[columnName]["Median"]=dataset[columnName].median()
        descriptive[columnName]["Mode"]=dataset[columnName].mode()[0]
        descriptive[columnName]["Min"]=dataset[columnName].min()
        descriptive[columnName]["Max"]=dataset[columnName].max()
        descriptive[columnName]["25%"]=dataset.describe()[columnName]["25%"]
        descriptive[columnName]["50%"]=dataset.describe()[columnName]["50%"]
        descriptive[columnName]["75%"]=dataset.describe()[columnName]["75%"]
        descriptive[columnName]["100%"]=dataset.describe()[columnName]["max"]
        descriptive[columnName]["IQR"]=descriptive[columnName]["75%"]-descriptive[columnName]["25%"]
        descriptive[columnName]["1.5IQR"]=1.5*descriptive[columnName]["IQR"]
        descriptive[columnName]["LesserRange"]=descriptive[columnName]["25%"]- descriptive[columnName]["1.5IQR"]
        descriptive[columnName]["GreaterRange"]=descriptive[columnName]["75%"]+ descriptive[columnName]["1.5IQR"]
        descriptive[columnName]["Skewness"]=dataset[columnName].skew()
        descriptive[columnName]["Kurtosis"]=dataset[columnName].kurtosis()
    return descriptive

Univariate(dataset,quan)
````

A frequency table shows how often different values appear in your data. It helps you understand patterns and distributions.

### Key Components:
1. **Unique Values** - Each different item in your data
2. **Frequency** - How many times each value appears
3. **Relative Frequency** - Percentage of total for each value
4. **Cumulative Sum** - Running total of percentages


## Understanding Frequency Concepts

### 1. Types of Frequency Measures

| Measure | Description | Formula | When to Use |
|---------|-------------|---------|-------------|
| **Frequency** | Raw count of each unique value | Count occurrences | Understanding absolute distribution |
| **Relative Frequency** | Proportion of each value in dataset | Frequency/Total observations | Comparing distributions with different sample sizes |
| **Cumulative Frequency** | Running total of frequencies | Sum of all previous frequencies | Analyzing data distribution patterns |
| **Cumulative Relative Frequency** | Running total of relative frequencies | Sum of all previous relative frequencies | Determining percentiles and quartiles |

### 2. Types of Variables

| Variable Type | Frequency Table Use | Example |
|--------------|---------------------|---------|
| **Categorical** | Most common use case | Gender, Product Categories |
| **Ordinal** | Shows ordered relationships | Education Levels, Satisfaction Ratings |
| **Discrete Numerical** | Works well with limited unique values | Number of children, Shoe sizes |
| **Continuous Numerical** | Often requires binning first | Age groups, Income ranges |

# Skewness and Kurtosis in Data

## Definitions

### Skewness
Skewness measures the **asymmetry** of a distribution:
- **Positive skew**: Right tail is longer (mean > median)
- **Negative skew**: Left tail is longer (mean < median)
- **Zero skew**: Perfectly symmetrical (mean = median)

### Kurtosis
Kurtosis measures the **tailedness** (peakedness and tail thickness) of a distribution:
- **High kurtosis**: Heavy tails and sharp peak (more outliers)
- **Low kurtosis**: Light tails and flatter peak (fewer outliers)
- **Normal distribution** has kurtosis ≈ 3 (often reported as "excess kurtosis" = 0)

````python
def Univariate(dataset,quan):
    descriptive=pd.DataFrame(index=["Mean","Median","Mode","Min","Max","25%","50%","75%","100%","IQR",
                               "1.5IQR","LesserRange","GreaterRange","Skewness","Kurtosis"],columns=quan)
    
    for columnName in quan:
        descriptive[columnName]["Mean"]=dataset[columnName].mean()
        descriptive[columnName]["Median"]=dataset[columnName].median()
        descriptive[columnName]["Mode"]=dataset[columnName].mode()[0]
        descriptive[columnName]["Min"]=dataset[columnName].min()
        descriptive[columnName]["Max"]=dataset[columnName].max()
        descriptive[columnName]["25%"]=dataset.describe()[columnName]["25%"]
        descriptive[columnName]["50%"]=dataset.describe()[columnName]["50%"]
        descriptive[columnName]["75%"]=dataset.describe()[columnName]["75%"]
        descriptive[columnName]["100%"]=dataset.describe()[columnName]["max"]
        descriptive[columnName]["IQR"]=descriptive[columnName]["75%"]-descriptive[columnName]["25%"]
        descriptive[columnName]["1.5IQR"]=1.5*descriptive[columnName]["IQR"]
        descriptive[columnName]["LesserRange"]=descriptive[columnName]["25%"]- descriptive[columnName]["1.5IQR"]
        descriptive[columnName]["GreaterRange"]=descriptive[columnName]["75%"]+ descriptive[columnName]["1.5IQR"]
        descriptive[columnName]["Skewness"]=dataset[columnName].skew()
        descriptive[columnName]["Kurtosis"]=dataset[columnName].kurtosis()
    return descriptive

Univariate(dataset,quan)
````

### Understanding Variance and Standard Deviation

````python
def Univariate(dataset,quan):
    descriptive=pd.DataFrame(index=["Mean","Median","Mode","Min","Max","25%","50%","75%","100%","IQR",
                               "1.5IQR","LesserRange","GreaterRange","Skewness","Kurtosis", "Var","Std"],columns=quan)
    for columnName in quan:
        descriptive[columnName]["Mean"]=dataset[columnName].mean()
        descriptive[columnName]["Median"]=dataset[columnName].median()
        descriptive[columnName]["Mode"]=dataset[columnName].mode()[0]
        descriptive[columnName]["Min"]=dataset[columnName].min()
        descriptive[columnName]["Max"]=dataset[columnName].max()
        descriptive[columnName]["25%"]=dataset.describe()[columnName]["25%"]
        descriptive[columnName]["50%"]=dataset.describe()[columnName]["50%"]
        descriptive[columnName]["75%"]=dataset.describe()[columnName]["75%"]
        descriptive[columnName]["100%"]=dataset.describe()[columnName]["max"]
        descriptive[columnName]["IQR"]=descriptive[columnName]["75%"]-descriptive[columnName]["25%"]
        descriptive[columnName]["1.5IQR"]=1.5*descriptive[columnName]["IQR"]
        descriptive[columnName]["LesserRange"]=descriptive[columnName]["25%"]- descriptive[columnName]["1.5IQR"]
        descriptive[columnName]["GreaterRange"]=descriptive[columnName]["75%"]+ descriptive[columnName]["1.5IQR"]
        descriptive[columnName]["Skewness"]=dataset[columnName].skew()
        descriptive[columnName]["Kurtosis"]=dataset[columnName].kurtosis()
        descriptive[columnName]["Var"]=dataset[columnName].var()
        descriptive[columnName]["Std"]=dataset[columnName].std()
    return descriptive
````


### Definitions

### Variance (σ²)
**Measures**: Spread of data points around the mean  

### Standard Deviation (σ)

**Measures**: Average distance from the mean (in original units)  

### Key Differences

| Characteristic | Variance | Standard Deviation |
|---------------|----------|---------------------|
| Units | Squared units (e.g., $²) | Original units (e.g., $) |
| Interpretation | Harder to interpret directly | More intuitive |
| Sensitivity | More sensitive to outliers | Less sensitive |

### When to Use Each

- **Use Variance** when:
  - Performing statistical tests (ANOVA, etc.)
  - Working with squared units in calculations

- **Use Standard Deviation** when:
  - Communicating results to non-technical audiences
  - Understanding real-world spread in original units

### **Normal Distribution Breakdown (Emprical Rule)** or Called Bell Shaped, Normal, Gaussian Distributions.

Your diagram shows a classic **bell curve** with standard deviation (σ) intervals from the mean (μ). Here's the complete explanation:

---

#### **1. Central Zone (μ ± 1σ)**
- **Range**: μ - 1σ to μ + 1σ  
- **Probability**: **68.2%** of data  
  (Shown as 34.1% + 34.1% in your diagram)  
- **Example**: If μ = 100 and σ = 15:  
  - 68% of values fall between **85-115**

#### **2. Intermediate Zone (μ ± 2σ)**
- **Range**: μ - 2σ to μ + 2σ  
- **Probability**: **95.4%** of data  
  (68.2% + 13.6% + 13.6%)  
- **Example**:  
  - 95% of values fall between **70-130**

#### **3. Outer Zone (μ ± 3σ)**
- **Range**: μ - 3σ to μ + 3σ  
- **Probability**: **99.7%** of data  
  (95.4% + 2.1% + 2.1% + 0.1% + 0.1%)  
- **Example**:  
  - 99.7% of values fall between **55-145**

---

### **Key Areas in Your Diagram**
| Label | Interpretation | Probability | Zone |
|-------|-----------------|-------------|------|
| μ | Mean (center) | N/A | Center |
| -30, -20, -10 | σ intervals (e.g., -3σ, -2σ, -1σ) | N/A | Left Tail |
| 0.1%, 2.1%, 13.6% | Right-tail probabilities | Cumulative beyond 1σ/2σ/3σ | Right Tail |
| 34.1% | Half of 68% zone | 34.1% per side | μ±1σ |

---

### **Practical Implications**
1. **Quality Control**:  
   - If a process output is normally distributed with μ = 100, σ = 5:  
     - **Red Flag**: Any value <85 (μ - 3σ) or >115 (μ + 3σ) is extremely rare (<0.3% chance).

2. **Grading Systems**:  
   - In exams with μ = 75, σ = 10:  
     - A grade (top 2.1%) ≈ score >95 (μ + 2σ)

3. **Finance**:  
   - Stock returns often follow this pattern:  
     - 95% of daily returns within ±2σ of average

---

### **Mathematical Confirmation**
For a true normal distribution:
```math
P(μ - σ ≤ X ≤ μ + σ) ≈ 68.27\% \\
P(μ - 2σ ≤ X ≤ μ + 2σ) ≈ 95.45\% \\
P(μ - 3σ ≤ X ≤ μ + 3σ) ≈ 99.73\%
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-3, 3, 1000)
y = norm.pdf(x, 0, 1)  # μ=0, σ=1

plt.figure(figsize=(10,5))
plt.plot(x, y, 'k-')
plt.fill_between(x, y, where=(x >= -1) & (x <= 1), color='blue', alpha=0.3, label='68% (μ±1σ)')
plt.fill_between(x, y, where=(x >= -2) & (x <= 2), color='green', alpha=0.2, label='95% (μ±2σ)')
plt.fill_between(x, y, where=(x >= -3) & (x <= 3), color='red', alpha=0.1, label='99.7% (μ±3σ)')
plt.legend()
plt.title("Standard Normal Distribution")
plt.xlabel("Standard Deviations (σ)")
plt.ylabel("Probability Density")
plt.show()
```

### **Common Misconceptions**
1. **All bell curves are normal**: Only true if they follow the 68-95-99.7 rule exactly.  
2. **Outliers beyond 3σ don't exist**: They're rare but possible (0.3% probability).  
3. **Data must be normal**: Many real-world datasets are only approximately normal.
